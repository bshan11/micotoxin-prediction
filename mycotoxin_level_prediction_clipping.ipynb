## Apply Clipping

**Why Clipping?**
- Clipping is a safer alternative to removing outliers, especially when you want to preserve data while handling extreme values.
- It improves stability and ensures models learn real patterns instead of noise.

```python
# Compute IQR for the target variable
Q1_target = df_cleaned["vomitoxin_ppb"].quantile(0.25)
Q3_target = df_cleaned["vomitoxin_ppb"].quantile(0.75)
IQR_target = Q3_target - Q1_target

# Define clipping bounds
lower_bound_target = Q1_target - 1.5 * IQR_target
upper_bound_target = Q3_target + 1.5 * IQR_target

# Apply clipping
df_cleaned["vomitoxin_ppb"] = df_cleaned["vomitoxin_ppb"].clip(lower=lower_bound_target, upper=upper_bound_target)

# Check new statistics after clipping
df_cleaned["vomitoxin_ppb"].describe()
```

```python
# Plot distribution before and after clipping
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Before Clipping (Original Distribution)
sns.histplot(df["vomitoxin_ppb"], bins=30, kde=True, ax=axes[0], color="red")
axes[0].set_title("Before Clipping: vomitoxin_ppb Distribution")
axes[0].set_xlabel("vomitoxin_ppb")
axes[0].set_ylabel("Frequency")

# After Clipping (Modified Distribution)
sns.histplot(df_cleaned["vomitoxin_ppb"], bins=30, kde=True, ax=axes[1], color="blue")
axes[1].set_title("After Clipping: vomitoxin_ppb Distribution")
axes[1].set_xlabel("vomitoxin_ppb")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## 2. Dimensionality Reduction

```python
# Apply PCA to spectral features
pca = PCA()
principal_components = pca.fit_transform(df_cleaned[feature_columns])

# Explained variance ratio
explained_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot cumulative explained variance with orange markers
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', 
         markerfacecolor='orange', linestyle='-')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA: Explained Variance vs. Number of Components")
plt.grid(True)
plt.show()

# Find number of components that explain at least 95% variance
num_components = np.argmax(explained_variance >= 0.95) + 1
num_components
```

```python
# Get explained variance for the top principal components
top_pcs_variance = pca.explained_variance_ratio_[:5]  # First 5 components

# Create a dataframe for better visualization
variance_df = pd.DataFrame({
    "Principal Component": [f"PC{i+1}" for i in range(5)],
    "Explained Variance Ratio": top_pcs_variance
})

# Display the variance explained by top 5 principal components
variance_df
```

## PCA Results ✅
- The cumulative explained variance plot shows how many principal components are needed to retain most of the data's information.
- **5** principal components explain at least **95%** of the variance in the dataset.
- **PC1** alone explains **85.24%** of the variance, meaning most spectral variations are captured in just one component.
- **PC2** adds another **5.99%**, helping to capture additional patterns.
- Together, **PC1–PC5** explain over **95%** of the variance, justifying our selection of 5 components.
